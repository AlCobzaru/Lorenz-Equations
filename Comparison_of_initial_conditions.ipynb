When comparing $r=25$ for a slightly different initial value (5 and 5.1), it is very clear that their trajectory deviate at around the 8 second mark.

When comparting $r=25$ to $r=25.1/25,001/25.00001$, the trajectories also deviate, only at increasingly further times. For 25 versus 25.1, the trajectories deviate around the 10th second.
