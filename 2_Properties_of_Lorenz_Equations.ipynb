# Properties of Lorenz Equations

**1) Symmetry**

The Lorenz equations exhibit symmetry with respect to x and y. In other words,

$ f(x, y) = f(-x,-y)$

**Proof**
Let's write the system of equations for -x and -y and simplify them (Note, the sign of z is unchanging because the symmetry does not apply to it):

For $ \frac{dx}{dt} $

$ -\frac{dx}{dt} = σ(-y - (-x)) $

$ -\frac{dx}{dt} = -σ(y - x)$

**$ \frac{dx}{dt} = σ(y - x)$ (1)**

For $ \frac{dy}{dt} $

$ -\frac{dy}{dt} = -x(\rho - z) - (-y) $

$ -\frac{dy}{dt} = -(x(\rho - z) - y) $

$ \frac{dy}{dt} = x(\rho - z) - y $ **(2)**

For $ \frac{dz}{dt} $

$ \frac{dz}{dt} = (-x)(-y) - \beta z $

$ \frac{dz}{dt} = xy - \beta z $ **(3)**

QED

This means that any set of solutions has a symmetric solution to it.

**2) Nonlinearity**

The Lorenz equations are not a set of linear differential equations, as they contain two linearities. In equation (2), -xz is not linear. In equation (3), xy is not linear.

**3) Volume dissipation**

As time advances, the lorenz system approaches a set of solution that has no volume. As a matter of fact, it approaches the zero volume exponentially fast.

**Proof**
Let's first study how to calculate the evolution of the volume of a sample object, which we will use to determine what happens in the Lorenz equations.

In the following image, an amorphous object with a given volume is changing shape. At the black mark, the infinitestimal surface area is moving with outward velocity **v** with respect to the norm at that surface area denoted **n**. The normal outward velocity of a function x (x, y, z) is the dot product between v and n)
Volume dissipation.jpg

For a small interval of time **dt**, the surface area will expand with a volume of its area time the depths:

$ V(dt) = (v	• n)dtdA $

The total volume is simply the sum of the volume at initial t and all of the small patches of volume of every dA. Integration is required to find the solution.

$$ V(t + dt) = V(t) + \iint_S (v	• n)dtdA $$

dt can be taken out of the integral, and by isolating the integral, the following equation is obtained:


$$ \frac{V(t + dt) - V(t)}{dt} = \iint_S (v	• n)dA $$

The left-hand side is simply the definition of the derivative. Thus, using the divergence theorem from vector calculus, this equation can be simplified in the following form:


$$ \frac{dV}{dt} = \iiint_V (∇ • v)dV $$

$$ \frac{dV}{dt} = (∇ • v) $$

Having found the function determining how the rate of the volume changes, we simply have to compute it in the Lorenz equations. **x** is the system of Lorenz equations. v is the set of all three rates of change of the unkowns x, y and z.

$$ ∇ • v = \frac{∂x'}{∂x} + \frac{∂y'}{∂y} +  \frac{∂z'}{∂z} $$

$$ ∇ • v = \frac{∂(σ(y - x))}{∂x} + \frac{∂(x(\rho - z) - y)}{∂y} +  \frac{∂(xy−βz)}{∂z} $$

$$ ∇ • v = -σ -1 -β $$

The divergence of this system of equations is constant, which can be simplified to:

$$ \frac{dV}{dt} = -σ -1 -β $$

Once this separable differential equation is solved, the function of the Volume is equal to the following:

$$ V = V(0)e^{(-σ-1-β)t} $$

Since the canonical values of sigma and beta are positive, the volume of the system is ever decreasing. This imposes a condition on all the solutions of the Lorenz system, meaning that the attractors can only be points, limit cycles, or strange attractors. Because the volume is ever decreasing, this means that there cannot be any repeller on the system, as a repeller inherently generates a volume.

**4) Critical Points (Fixed points)**

Let's attempt to find critical points of the Lorenz system. First, set the three equations to be equal to 0 and derive conditions:

For $ \frac{dx}{dt} $

$ 0 = σ(y - x)$

$ y = x $ (1)


For $ \frac{dy}{dt} $

$ 0 = x(\rho - z) - y $ (2)

For $ \frac{dz}{dt} $

$ 0 = xy - \beta z $

$ 0 = x^2 - \beta z $ (using (1)) (3)

Since every term is a multiple of a function x, y, z, then an obvious solution is for a point where x, y, z = 0, that is P(0,0,0).

If we set $ z = \rho - 1 $, equation (2) simplifies to:

$ 0 = x(\rho - (\rho - 1)) - y $

$ 0 = x - y $

Which is the same as equation (1). This leaves equation (3), which, when simplified, gives two new values of critical points:

$ x^2 = βz $

$ x =± \sqrt{β(ρ -1 )} $

Setting the paramter β to its canonical value of $ \frac{8}{3} $, the equation simplifies to:

$ x =± \sqrt{\frac{8}{3} (ρ -1 )} $

These two new critical points only exist when ρ > 1, meaning that the set of critical points begin with only one point, then suddenly, as r increases and reaches a bit more than 1, two new critical points, attractors, suddenly exist. Here is the notation for the two other critical points:

$ P( \sqrt{\frac{8}{3} (ρ -1 )},\sqrt{\frac{8}{3} (ρ -1 )},ρ -1) $

$ P( -\sqrt{\frac{8}{3} (ρ -1 )},-\sqrt{\frac{8}{3} (ρ -1 )},ρ -1) $
